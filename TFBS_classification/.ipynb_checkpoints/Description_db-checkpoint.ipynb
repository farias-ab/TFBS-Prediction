```python
import pandas as pd
```

# RegulonDB

```python
reg_db = pd.read_csv('./dataset/regulon_db.csv', sep='\t', header=None)
reg_db
```

```python
# Verificando duplicatas
dup_reg = reg_db[reg_db.duplicated()]
dup_reg
```

```python
reg_nodup = reg_db.drop_duplicates()
reg_nodup
```

# CollecTF

```python
col_db = pd.read_csv('./dataset/collectf_db.csv', sep='\t', header=None)
col_db
```

```python
# Verificando duplicatas
dup_col = col_db[col_db.duplicated()]
dup_col
```

```python
col_nodup = col_db.drop_duplicates()
col_nodup
```

# Regulon + CollecTF

```python
df_reg_col = pd.concat([reg_nodup, col_nodup], ignore_index=True)
df_reg_col
```

```python
# Verificando duplicatas
dup_reg_col = df_reg_col[df_reg_col.duplicated()]
dup_reg_col
```

```python
nodup_reg_col = df_reg_col.drop_duplicates()
nodup_reg_col
```

```python
def calcular_numero_de_colunas(row):
    return row.count()
```

```python
nodup_reg_col['numero_colunas'] = nodup_reg_col.apply(calcular_numero_de_colunas, axis=1)
tabela_frequencia = nodup_reg_col['numero_colunas'].value_counts().sort_index()
tabela_frequencia
```

```python
nodup_reg_col
```

# Preparing our dataset to ML

```python
nodup_reg_col = nodup_reg_col.drop('numero_colunas', axis=1)
nodup_reg_col
```

```python
df_8 = nodup_reg_col[nodup_reg_col.apply(lambda row: len(row.dropna()) == 8, axis=1)]
df_8 = df_8.dropna(axis=1)
df_8
```

```python
df_8.to_csv('reg_col_8bps.csv', index=False)
```
