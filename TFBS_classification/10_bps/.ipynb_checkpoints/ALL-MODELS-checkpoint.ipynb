```python
# Importando as bibliotecas necessárias
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, roc_auc_score
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.linear_model import SGDClassifier
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('dataset_bps_9.csv', sep=',')
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Definindo o random state e a proporção de split
random_state = 42
test_size = 0.20

# Dividindo os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
```

```python
# Lista de classificadores
classifiers = [
    ("AdaBoost", AdaBoostClassifier(random_state=random_state)),
    ("Decision Tree", DecisionTreeClassifier(random_state=random_state)),
    ("Gradient Boosting", GradientBoostingClassifier(random_state=random_state)),
    ("Random Forest", RandomForestClassifier(random_state=random_state)),
    ("SGD Classifier", SGDClassifier(loss='log', random_state=random_state)),
    ("SVC", SVC(probability=True, random_state=random_state))
]
```

```python
# Inicializando listas para armazenar taxas de falsos positivos (FPR) e taxas de verdadeiros positivos (TPR)
fpr_list = []
tpr_list = []
roc_auc_list = []
```

```python
# Iterando sobre os classificadores
for name, classifier in classifiers:
    # Treinando o modelo
    classifier.fit(X_train, y_train)
    
    # Fazendo previsões no conjunto de teste
    y_pred = classifier.predict(X_test)
    
    # Calculando as métricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Imprimindo as métricas
    print(f"Metrics for {name}:")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")
    
    # Realizando a validação cruzada com 10 folds
    cv_scores = cross_val_score(classifier, X, y, cv=10)
    print(f"Cross-Validation Mean Accuracy: {cv_scores.mean():.2f}\n")
```

```python
# Iterando sobre os classificadores
for name, classifier in classifiers:
    # Treinando o modelo
    classifier.fit(X_train, y_train)
    
    # Calculando a curva ROC e a área sob a curva (AUC)
    y_scores = classifier.predict_proba(X_test)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, y_scores)
    roc_auc = roc_auc_score(y_test, y_scores)
    
    # Armazenando os valores para plotagem posterior
    fpr_list.append(fpr)
    tpr_list.append(tpr)
    roc_auc_list.append(roc_auc)
    
    # Imprimindo a área sob a curva para cada modelo
    print(f"AUC for {name}: {roc_auc:.2f}")

# Criando o gráfico ROC para comparar todos os modelos
plt.figure(figsize=(10, 7))

for i in range(len(classifiers)):
    plt.plot(fpr_list[i], tpr_list[i], label=f'{classifiers[i][0]} (AUC = {roc_auc_list[i]:.2f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva ROC Comparativa para Modelos de Classificação')
plt.legend(loc='lower right')
plt.show()
```

```python
# Criando um DataFrame com os dados das curvas ROC
roc_data = pd.DataFrame({
    'Classifier': [classifiers[i][0] for i in range(len(classifiers))],
    'FPR': fpr_list,
    'TPR': tpr_list,
    'AUC': roc_auc_list
})

# Salvando os dados em um arquivo CSV
roc_data.to_csv('ALL_curva_roc_data.csv', index=False)
```
